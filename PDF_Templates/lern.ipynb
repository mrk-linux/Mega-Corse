```python
from fpdf import FPDF
import pandas as pd


pdf = FPDF(orientation= "P", unit= "mm", format= "A4")
pdf.set_auto_page_break(auto= False, margin= 0) #Set auto page break mode and triggering margin

df = pd.read_csv(r"D:\porgram\code\mega corse\PDF_Templates\topics.csv") 
#Read a comma-separated values (csv) file into DataFrame.

for index, row in df.iterrows():
    #Iterate over DataFrame rows as (index, Series) pairs.
    pdf.add_page() #Start a new page
    pdf.set_font(family= "Times", style= "B", size= 24)
    #Select a font; size given in points
    pdf.set_text_color(100,100, 100) #Set color for text
    pdf.cell(w= 0, h= 12, txt= row["Topic"], align= "L", ln= 1)
    pdf.line(10, 21 ,200, 21) #Draw a line

    for y in range(20, 298, 10):
        pdf.line(10, y , 200, y)
    #set the footer
    pdf.ln(265)
    pdf.set_font(family= "Times", style= "I", size= 8)
    pdf.set_text_color(180,180, 180)
    pdf.cell(w= 0, h= 10,txt= row["Topic"], align= "r")

    for i in range(row["Pages"] - 1):
        pdf.add_page()
        pdf.ln(277)#Line Feed; default value is last cell height
        pdf.set_font(family= "Times", style= "I", size= 8)
        pdf.set_text_color(180,180, 180)
        pdf.cell(w= 0, h= 10,txt= row["Topic"], align= "r")
        for y in range(20, 298, 10):
            pdf.line(10, y , 200, y)

pdf.output("output.pdf")
```
