```python
import P_functions
import FreeSimpleGUI as sg
import time


#sg.theme("") شما می توانید تم مورد نظر خودتون رو از اینترنت پیدا کنید یا دانلود کنید و داخل پرانتز قرار دهید.

clock = sg.Text("",key="clock",text_color="purple")
"""
Text - یک متن را در یک پنچره ای نشان میدهد.  
معمولا از این متد برای تک خط استفاده می شود.  
با این حال می توان چندین خط خط را نیز نوشت . 
اگر چندین خط دارید نمیتوان ان را اسکرول کرد.
"""
label = sg.Text("type in a to-do")
input_box = sg.InputText(tooltip="enter to-do", key="todo")
add_button = sg.Button("add")
"""
Button Element - یک دکمه پیشفرض درست می کند.
یک شرت کات برای اجرای یک فرایند, سرچ فایل, ... هرکاری که یک دکمه بتواند بکند.
"""
list_box = sg.Listbox(values= P_functions.read_todos(), 
                    key="todos", enable_events= True, 
                    size=[45, 10])
"""
A List Box.  
یک لیست خصوصی از مقادیر برای کاربر تا بتواند از بین انها یکی را انتخاب کند.   
می تواند لیست انتخاب شده را برگرداند
اگر  window.read() وجود داشته باشد.
"""
edit_button = sg.Button("edit")

complete_button=sg.Button("complete")

exit_button = sg.Button("exit")

window = sg.Window("my to-do app",
                    layout=[[clock],
                            [label],[input_box, add_button],
                            [list_box, [edit_button,complete_button]],
                            [exit_button]],
                    font=("helvetica", 14))

while True:
    event, values = window.read(timeout=500)
    window["clock"].update(value= time.strftime("%b/%d/%Y\n%p %I:%M"))
    match event:

        case "add":
            todos = P_functions.read_todos()
            new_todo = values["todo"] + "\n"
            todos.append(new_todo)
            P_functions.write_todos(todos)
            window["todos"].update(values=todos)

        case "edit":
            try:
                todo_to_edit = values["todos"][0]
                new_todo = values["todo"]
                todos = P_functions.read_todos()
                index = todos.index(todo_to_edit)
                todos[index] = new_todo
                P_functions.write_todos(todos)
                window["todos"].update(values=todos)
            except IndexError:
                sg.popup("Please select an item first",font=("helvetica", 14))
            """
            Popup - یک پنجره پاپ‌آپ را نمایش دهید با هر تعداد پارامتر دلخواهی که می‌خواهید اضافه کنید.  
            این معادل GUI پرینت هست
            این همچنین عالی است برای «توقف» جریان برنامه تا کاربر بتواند پیام‌ های خطا را بخواند.
            """
        case "complete":
            try:
                todo_to_complete = values['todos'][0]
                todos = P_functions.read_todos()
                todos.remove(todo_to_complete)
                P_functions.write_todos(todos)
                window["todos"].Update(values=todos)
                window["todo"].Update(value="")
            except IndexError:
                sg.popup("Please select an item first",font=("helvetica", 14))

        case"exit":
            break

        case "todos":
            window['todo'].update(value=values["todos"][0])

        case sg.WIN_CLOSED:
            break
window.close()
```
