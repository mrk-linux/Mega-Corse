```python
from fpdf import FPDF
import pandas as pd


pdf = FPDF(orientation= "P", unit= "mm", format= "A4")
pdf.set_auto_page_break(auto= False, margin= 0)  # غیرفعال کردن حالت صفحه‌شکن خودکار و تنظیم حاشیه

df = pd.read_csv(r"mega corse\persian mega corse\فرمت_pdf\topics.csv") 
# یک فایل با مقادیر جدا شده با کاما (csv) را در دیتا فریم (DataFrame) بخوان.

for index, row in df.iterrows():
    # پیمایش ردیف‌های دیتا فریم به صورت (ایندکس، سری)ها
    pdf.add_page()  # شروع صفحه جدید
    pdf.set_font(family= "Times", style= "B", size= 24)
    # انتخاب قلم؛ سایز بر حسب نقطه
    pdf.set_text_color(100,100, 100)  # تنظیم رنگ متن
    pdf.cell(w= 0, h= 12, txt= row["Topic"], align= "L", ln= 1)
    pdf.line(10, 21 ,200, 21)  # رسم خط

    for y in range(20, 298, 10):
        pdf.line(10, y , 200, y)
    # تنظیم فوتر (پاورقی/پایین صفحه)
    pdf.ln(265)
    pdf.set_font(family= "Times", style= "I", size= 8)
    pdf.set_text_color(180,180, 180)
    pdf.cell(w= 0, h= 10,txt= row["Topic"], align= "r")

    for i in range(row["Pages"] - 1):
        pdf.add_page()
        pdf.ln(277)  # تغذیه خط؛ مقدار پیش‌فرض برابر ارتفاع آخرین سلول است
        pdf.set_font(family= "Times", style= "I", size= 8)
        pdf.set_text_color(180,180, 180)
        pdf.cell(w= 0, h= 10,txt= row["Topic"], align= "r")
        for y in range(20, 298, 10):
            pdf.line(10, y , 200, y)

pdf.output("output.pdf")
```
